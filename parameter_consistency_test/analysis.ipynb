```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
shapes = []
exps = []
for i in range(1, 76):
    fname = f"frame_{str(i).zfill(3)}"
    shapes.append(np.load("results/" + fname + "_np_shape.npy")[0])
    exps.append(np.load("results/" + fname + "_np_exp.npy")[0])
```

```python
shapes_mat = np.array(shapes)
exps_mat = np.array(exps)
```

```python
plt.plot(shapes_mat[:, 0])
plt.show()
```

```python
plt.plot(exps_mat[:, 0])
plt.show()
```

```python
shapes_mean = np.mean(shapes_mat, axis=0)
shapes_mean
```

```python
shapes_std = np.std(shapes_mat, axis=0)
shapes_std
```

```python
exps_mean = np.mean(exps_mat, axis=0)
exps_mean
```

```python
exps_std = np.std(exps_mat, axis=0)
exps_std
```
